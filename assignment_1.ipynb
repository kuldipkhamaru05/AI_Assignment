```python
import pymongo
import pandas as pd
myclient = pymongo.MongoClient()    
myCollection = myclient.python_db.samples
#for displaying the raw data from mongo db
# for doc in myCollection.find():
#     print(doc)
#displaying the raw data in dataframe
pd.DataFrame(list(myCollection.find()))
```

```python
#get sample data from db
getSampleData = pd.DataFrame(list(myCollection.find()))
#save those data frame data to csv
getSampleData.to_csv('saveSampleData.csv', encoding='utf-8')
```

```python
#reading that newly created csv file via panda
pd.read_csv('saveSampleData.csv')
```

```python
#create another extra column apart from existing data set and via python code only and in that column
#generate 100 random data
#my column name is "new_column_added" and over there I just multiplied 2 with the id value from DB
df = pd.DataFrame(list(myCollection.find()))
df['new_column_added'] = df['id'] * 2
df[['_id', 'userId', 'id', 'new_column_added', 'title', 'body', 'createdAt', 'updatedAt', '__v']].head(100)
```

```python
#print top 5 rows
df = pd.DataFrame(list(myCollection.find()))
df['new_column_added'] = df['id'] * 2
df[['_id', 'userId', 'id', 'new_column_added', 'title', 'body', 'createdAt', 'updatedAt', '__v']].head()
```

```python
#print what dimensional matrix this dataframe
df = pd.DataFrame(list(myCollection.find()))
df['new_column_added'] = df['id'] * 2
df[['_id', 'userId', 'id', 'new_column_added', 'title', 'body', 'createdAt', 'updatedAt', '__v']].head()
df.ndim
```

```python
#Transpose the above dataframe
df = pd.DataFrame(list(myCollection.find()))
df['new_column_added'] = df['id'] * 2
df[['_id', 'userId', 'id', 'new_column_added', 'title', 'body', 'createdAt', 'updatedAt', '__v']].head()
df.transpose()
```

```python
#after Transposing print what dimensional matrix this dataframe
df = pd.DataFrame(list(myCollection.find()))
df['new_column_added'] = df['id'] * 2
df[['_id', 'userId', 'id', 'new_column_added', 'title', 'body', 'createdAt', 'updatedAt', '__v']].head()
df.transpose()
df.ndim
```
